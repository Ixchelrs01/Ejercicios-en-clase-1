# Descripci칩n de las cantidades estad칤sticas




*   **Media**: es el valor promedio de un conjunto de datos , calculada como la suma del conjunto de valores dividida entre el n칰mero total de valores.


*   **Mediana**: es el valor que ocupa el lugar central de todos los datos cuando 칠stos est치n ordenados de menor a mayor.



*   **Moda**: es el valor que aparece con mayor frecuencia en un conjunto de datos


*   **Rango**:es un valor num칠rico que indica la diferencia entre el valor m치ximo y el m칤nimo de una poblaci칩n o muestra estad칤stica.



*   **Variancia**: mide en qu칠 medida se dispersan los valores de una muestra alrededor de un valor central.


*   **Desviaci칩n est치ndar**:es la medida de dispersi칩n que indica qu칠 tan dispersos est치n los datos con respecto a la media


*   **Error est치ndar**: Es la medida de la incertidumbre de la estad칤stica de prueba.












# Variables estad칤sticas del momento transverso del primer mu칩n

```python
#Histogramas y recolecci칩n de datos
#autor: Ixchel Ramirez Solano
#fecha: Diciembre 01, 2021

import numpy as np
import statistics as stat
from scipy import stats
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt

#Lectura del archivo

file= "http://opendata.cern.ch/record/5201/files/Dimuon_DoubleMu.csv"

particles= pd.read_csv(file)

data= np.genfromtxt(file,delimiter=",", names=["Run","Event", "type1", "E1", "px1", "py1", "pz1","pt1", "eta1", "phi", "Q1", "type2", "E2", "px2", "py2", "pz2","pt2", "eta2", "phi2", "Q2","M"], skip_header=1)

#Seleccion de los datos entre 35 y 50 GeV
eventos_filtrados= data[data["pt1"]>=35]
eventos_filtrados= eventos_filtrados[eventos_filtrados["pt1"]<=50]

media = round(np.mean(eventos_filtrados["pt1"]),3)
mediana= round(np.median(eventos_filtrados["pt1"]),3)
moda= stats.mode(eventos_filtrados["pt1"])
rango=round(np.ptp(eventos_filtrados["pt1"]),3)
variancia= round(stat.pvariance(eventos_filtrados["pt1"]),3)
desviacion= round(stat.pstdev(eventos_filtrados["pt1"]),3)
error= round(stats.sem(eventos_filtrados["pt1"]),3)

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", round(float(moda[0]),3))
print("Rango:", rango)
print("Variancia:", variancia)
print("Desviaci칩n est치ndar:", desviacion)
print("Error est치ndar:", error)

#Creacion del histograma entre 35 a 50 GeV
grafica= px.histogram(data["pt1"],data["pt1"],nbins=1800)
grafica.update_layout(title=r"Momento transverso del primer mu칩n",xaxis_title=r"Momento transverso [GeV]", yaxis_title="Eventos")
grafica.update_layout(xaxis_range=[35,50])
grafica.update_layout(yaxis_range=[0,180])
grafica.add_annotation(x=37, y=165, text="media="+str(media), showarrow=False)
grafica.add_annotation(x=40, y=165, text="Desviaci칩n est치ndar="+str(desviacion), showarrow=False)
grafica.add_annotation(x=43, y=165, text="Error est치ndar="+str(error), showarrow=False)
grafica.show()

```
