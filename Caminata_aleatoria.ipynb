```python
#
#
#

import random
from matplotlib import pyplot
import matplotlib.pyplot as plt

def movimientoX():
  deltaX= random.randint(-100,100)
  return deltaX/100

def movimientoY():
  deltaY= random.randint(-100,100)
  return deltaY/100

print(movimientoX() ,movimientoY())


desplazamientoX=[]
desplazamientoY=[]

Xinicial=0
Yinicial=0
counter= 0
desplazamientoX.append(Xinicial)
desplazamientoY.append(Yinicial)
x=Xinicial
y=Yinicial

NumerodeIteracion= int (input("numero de iteraciones"))


while counter<NumerodeIteracion:
  x+= movimientoX()
  y+= movimientoY()
  desplazamientoX.append(x)
  desplazamientoY.append(y)
  counter+=1

plt.plot(desplazamientoX, desplazamientoY)
plt.plot(x,y,marker="o",color="m", label="Punto final")
plt.plot(Xinicial,Yinicial,marker="o",color="r",label="Punto inicial")
plt.show()


```
